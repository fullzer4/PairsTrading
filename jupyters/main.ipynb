```python
import yfinance as yf
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

tickers = [
    "PETR4.SA",  # Petrobras PN
    "VALE3.SA",  # Vale ON
    "ITUB4.SA",  # Itaú Unibanco PN
    "BBDC4.SA",  # Bradesco PN
    "ABEV3.SA",  # Ambev ON
    "GGBR4.SA",  # Gerdau PN
    "CSNA3.SA",  # CSN ON
    "MGLU3.SA",  # Magazine Luiza ON
    "VIVT3.SA",  # Telefônica Brasil ON
    "B3SA3.SA",  # B3
    "BBAS3.SA",  # Banco do Brasil SA
    "BRKM5.SA",  # Braskem SA
    "ELET3.SA",  # Eletrobras ON
    "EMBR3.SA",  # Embraer SA
    "JBSS3.SA",  # JBS SA
    "LREN3.SA",  # Lojas Renner
    "PCAR3.SA",  # Grupo Pão de Açúcar
    "RADL3.SA",  # Raia Drogasil
    "SAPR11.SA", # Sanepar
    "WEGE3.SA",  # WEG SA
    "EQTL3.SA",  # Equatorial Energia
    "FLRY3.SA",  # Fleury SA
    "HYPE3.SA",  # Hypera Pharma
    "COGN3.SA",  # Cogna Educação (ex-Kroton)
    "NTCO3.SA",  # Natura
    "RENT3.SA",  # Localiza Rent a Car
    "SUZB3.SA",  # Suzano Papel e Celulose
    "IGTI11.SA"  # Iguatemi (ex-IGTA3.SA)
]
```

```python
data = yf.download(tickers, start="2018-01-01", end="2023-12-31")['Close']

data = data.fillna(method='ffill')

correlation_matrix = data.corr()
mask = (correlation_matrix > 0.75) & (correlation_matrix < 1)
filtered_corr = correlation_matrix[mask].fillna(0)

plt.figure(figsize=(12, 8))
sns.heatmap(filtered_corr, annot=True, cmap='coolwarm', fmt=".2f", linewidths=.5, vmin=0.75, vmax=1, cbar=False)
plt.title('Correlação entre Ações da B3')
plt.show()
```
